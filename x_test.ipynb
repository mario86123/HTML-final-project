```python
outer_with_encoding = pd.read_csv('/content/drive/MyDrive/HTML final project/outer_with_encoding.csv')
Test_IDs = pd.read_csv('/content/drive/MyDrive/HTML final project/Test_IDs.csv')
testing_data = pd.merge(outer_with_encoding, Test_IDs, on=["Customer ID"])

# deal with duplicates Customer data
testing_data = (testing_data.iloc[:, 1:-1].sort_values(['Customer ID'])
              .groupby(['Customer ID'])
              .apply(lambda x: x.ffill().bfill())
              .drop_duplicates())

# resort to original order
testing_data = testing_data.set_index('Customer ID')
testing_data = testing_data.reindex(index=Test_IDs['Customer ID'])
testing_data = testing_data.reset_index()

x_test = testing_data.drop('Customer ID', 1)
print(x_test.isna().sum())
```

```python
x_test.to_csv('/content/drive/My Drive/HTML final project/x_test.csv', index=False)
x_test
```
