```python
# read CSV
demographics = pd.read_csv('/content/drive/MyDrive/HTML final project/demographics.csv')

demographics['Gender'].replace(['Female', 'Male'], [0, 1],inplace = True)
demographics['Under 30'].replace(['No', 'Yes'], [0, 1],inplace = True)
demographics['Senior Citizen'].replace(['No', 'Yes'], [0, 1],inplace = True)
demographics['Married'].replace(['No', 'Yes'], [0, 1],inplace = True)
demographics['Dependents'].replace(['No', 'Yes'], [0, 1],inplace = True)

demographics
```

```python
print(demographics.isna().sum())
demographics.loc[demographics['Age'] < 30, ['Under 30']] = 1
demographics.loc[demographics['Age'] >= 30, ['Under 30']] = 0
demographics.loc[demographics['Age'] >= 65, ['Senior Citizen']] = 1
demographics.loc[demographics['Age'] < 65, ['Senior Citizen']] = 0
demographics.loc[demographics['Under 30'] == 1, ['Senior Citizen']] = 0
demographics.loc[demographics['Senior Citizen'] == 1, ['Under 30']] = 0
print(demographics.isna().sum())
```

```python
under_30 = demographics[(demographics['Under 30'] == 1.0) & demographics['Age'].notnull()]
under_30_median_age = under_30['Age'].median()
print("under_30_median: ", under_30_median_age)

senior = demographics[(demographics['Senior Citizen'] == 1.0) & demographics['Age'].notnull()]
senior_median_age = senior['Age'].median()
print("senior_median: ", senior_median_age)

between = demographics[(demographics['Under 30'] == 0.0) & (demographics['Senior Citizen'] == 0.0) & demographics['Age'].notnull()]
between_median_age = between['Age'].median()
print("between_median: ", between_median_age)

above_30 = demographics[(demographics['Under 30'] == 0.0) & demographics['Age'].notnull()]
above_30_median_age = above_30['Age'].median()
print("above_30_median: ", above_30_median_age)

under_65 = demographics[(demographics['Senior Citizen'] == 0.0) & demographics['Age'].notnull()]
under_65_median_age = under_65['Age'].median()
print("under_65_median: ", under_65_median_age)
```
